```python
import numpy as np
import matplotlib.pyplot as plt

x = [2, 5, 7, 10, 12]
y = [20, 10, 30 , 15, 5]

plt.plot(x, y, 'rx', label = 'x')

x = [10, 20, 30]
y = [4, 5, 3]
plt.plot(x, y, 'r', label = 'y')
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 1, 101)
y = x + np.random.random(len(x)) * x
plt.plot(x, y, 'r.')

A = np.vstack([x, np.ones(len(x))]).T

y = y[:, np.newaxis]

# answer = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), y)
answer = np.dot(np.linalg.pinv(A), y)

m = answer[0]
c = answer[1]

# print (y)
plt.plot(x, m*x + c)
plt.show()
```

```python
#return array yang isinya 100 data, dari 1-100. Semua datanya selisihnya sama
x = np.random.normal(5, 0.5, 100)

#np,random.normal(mean, standard deviation, jumlah data)
y = np.random.normal(5, 0.1, 100)/x
plt.plot(x, y, 'rx')

A = np.vstack([x, np.ones(len(x))]).T

y = y[:, np.newaxis]

answer = np.dot(np.linalg.pinv(A), y)

m = answer[0]
c = answer[1]

plt.plot(x, m*x + c)
plt.show()
```
