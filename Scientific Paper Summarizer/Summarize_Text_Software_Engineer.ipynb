```python
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
import numpy as np
import ipywidgets as widgets
from IPython.display import display, Audio
from gtts import gTTS

nltk.download('punkt')
nltk.download('stopwords')
```

```python
def summarize_text(text, num_sentences=3):
    # Split the text into sentences
    sentences = sent_tokenize(text)

    if len(sentences) <= num_sentences:
        return text

    vectorizer = TfidfVectorizer(stop_words=stopwords.words('english'))
    tfidf_matrix = vectorizer.fit_transform(sentences)
    sentence_scores = np.array(tfidf_matrix.sum(axis=1)).ravel()

    top_sentence_indices = sentence_scores.argsort()[-num_sentences:]
    top_sentence_indices.sort()

    # Generate the summary
    summary = ' '.join([sentences[i] for i in top_sentence_indices])
    return summary
```

```python
# Function to read text from a file
def read_text_from_file(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
    return text

# Function to display the summary in a text box and convert it to speech
def display_summary_and_speak(file_path, num_sentences=3):
    text_to_summarize = read_text_from_file(file_path)
    summary = summarize_text(text_to_summarize, num_sentences=num_sentences)

    summary_text_box = widgets.Textarea(
        value=summary,
        placeholder='Summary will be displayed here',
        description='Summary:',
        disabled=False,
        layout=widgets.Layout(width='100%', height='200px')
    )
    display(summary_text_box)
    tts = gTTS(summary, lang='en')
    tts.save('summary.mp3')

    # Play the audio
    display(Audio('summary.mp3', autoplay=True))

# Path to your text file
file_path = 'Text Test.txt'  # Replace with the path to your text file
display_summary_and_speak(file_path, num_sentences=3)
```
